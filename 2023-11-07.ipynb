```python
a = np.array([[[1, 2, 3], [2, 1, 3]],[[2, 1, 0], [2, 4, 1]]])
b = np.array([[[2, 1, 1], [2, 4, 0]],[[2, 1, 4], [2, 1, 3]]])
b1 = b.reshape(2, 3, 2)
#c = a @ b
print(b1)
print("----")
c = np.dot(a,b1)
print(c,c.shape)
print("----")
d= a @ b1
print(d,d.shape)
```
